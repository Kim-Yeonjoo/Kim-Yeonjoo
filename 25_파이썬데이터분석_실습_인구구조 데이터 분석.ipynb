# 데이터 전처리

```python
# 인덱스 열 불필요한 숫자 제거
df['행정구역'] = df['행정구역'].str.replace(r'\s*\(.*\)', '', regex=True)
df['행정구역']
```

```python
# 인덱스 열 제외 쉼표 제거
df.iloc[:, 1:] = df.iloc[:, 1:].replace(',', '', regex=True)
df.head()
```

```python
# 인덱스 열 제외 숫자형 변경
df_columns = df.columns[1:]
df[df_columns] = df[df_columns].apply(pd.to_numeric, errors='coerce').astype(float)
```

```python
df.dtypes
```

```python
df.head()
```

```python
#전라북도, 전북특별자치도 합치기
jeonbuk = df[df['행정구역'] == '전라북도']
jeonbuk_special = df[df['행정구역'] == '전북특별자치도']

combined = jeonbuk.iloc[0].copy()
combined['행정구역'] = '전북특별자치도'
for col in df.columns[1:]:  # '시도' 열을 제외한 모든 열에 대해
    combined[col] = jeonbuk.iloc[0][col] + jeonbuk_special.iloc[0][col]

df = df[df['행정구역'] != '전라북도']
df = df[df['행정구역'] != '전북특별자치도']

df.loc[19] = combined

df
```

## 현황 분석 (2024년 9월)

### 지역별 인구 비율

```python
#전국제외
df = df[df['행정구역'] != '전국']
#총인구 계산
total_population = df['2024년09월_총인구수'].sum()

# 지역별 인구 비율 계산
df['2024년09월_인구비율'] = (df['2024년09월_총인구수'] / total_population * 100).round(2)

df
```
