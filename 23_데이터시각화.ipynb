## 새로운 데이터 셋 로드
- 'tips' 데이터 셋, 팁, 청구서 전체금액, 일행수, 요일, 시간, 흡연여부, 성별

```python
tips = sns.load_dataset('tips')
tips.head()
```

## 2.1 일변량 그래프

### 히스토그램 hist()
- 연속형 변수  
- 구간의 "빈도수"를 시각화하는 그래프, x의 구간을 나누어 y는 빈도수

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.hist(data = tips, x='total_bill', bins=5) #bins : x 값 구간
fig.suptitle('Total Bill')
plt.show()
```

### 막대 countplot()
- 범주형 연속적 값 비교

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
sns.countplot(data = tips, x='day')
fig.suptitle('Day')
plt.show()
```

### 상자 boxplot()
- 연속형 변수  
- 통체로 분포 나타내는 그래프

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
sns.boxplot(data = tips, x='total_bill')
fig.suptitle('Total Bill')
plt.show()
```

## 2.2 이변량 그래프

### 산전도 scatter plot()
- 상관관계 확인

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.scatter( tips['total_bill'], tips['tip'] ) #(첫번째 변수, 두번째 변수)
fig.suptitle('총지불금액과 팁의 관계')
plt.show()
```

### 상자 boxplot()

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)

axes.boxplot(
    x=[  tips[tips['sex'] == 'Female' ]['tip'],
         tips[tips['sex'] == 'Male'   ]['tip']
      ], 
    labels=['Female','Male']
) 

fig.suptitle('성별과 팁의 관계')
plt.show()
```

## 2.3 다변량 그래프

```python
#map()함수에 사용
colors = {"Female": "#f1a340", "Male": "#998ec3"} # 성별에 따른 색상을 지정, 여성은 주황색, 남성은 보라색 
#16진수 코드는 https://www.w3schools.com/colors/colors_picker.asp 에서 확인 가능

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(data=tips,
              x='total_bill',
              y='tip',
              s=tips['size']**2*10,# size를 제곱하여 점의 크기를 키움, s는 점의 크기를 나타내는 매개변수
              c=tips['sex'].map(colors),# map()은 Series의 각 요소에 대해 특정 함수를 적용할 때 사용, c는 점의 색을 나타내는 매개변수, colors 딕셔너리에 따라 여성은 주황색, 남성은 보라색
              alpha=0.5) # alpha는 점의 투명도를 나타내는 매개변수

axes1.set_title('Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

scatter_plot.suptitle('Total Bill vs Tip')

plt.show()
```

# 3. Seaborn 라이브러리
- 기본적으로 x,y축을 만들어줌

```python
import seaborn as sns
```

## 3.1 일변량 그래프

### histplot()

```python
sns.histplot(data=tips, x='total_bill')
plt.show()
```

### kdeplot()

```python
sns.kdeplot(data=tips, x='total_bill')
plt.show()
```

### rugplot()

```python
sns.rugplot(data=tips, x='total_bill')

sns.histplot(data=tips, x='total_bill') # rugplot은 보통 다른 그래프와 함께 보여줌
plt.show()
```

### countplot()

```python
cnt, ax = plt.subplots() #supplots > tuple

sns.countplot(data=tips, x='day', palette='viridis') 
#palette: 정해진 디자인
#https://seaborn.pydata.org/tutorial/color_palettes.html

ax.set_title('요일별 건수')

plt.show()
```

## 3.2 이변량 그래프

### scatterplot()

```python
cnt, ax = plt.subplots() #supplots > tuple

sns.scatterplot(data=tips, x='total_bill', y='tip', ax=ax, hue='sex', palette='viridis') 
                                                         #hue:범례
ax.set_title('지불금액과 팁의 관계')

plt.show()
```

### regplot()

```python
cnt, ax = plt.subplots() #supplots > tuple
sns.regplot(data=tips, x='total_bill', y='tip', ax=ax)  #huge, palette 없음                                                      
ax.set_title('지불금액과 팁의 관계')
plt.show()
```

### jointplot()

```python
joint = sns.jointplot(data=tips, x='total_bill', y='tip')
joint.set_axis_labels(xlabel='TotalBill', ylabel='Tip')
joint.figure.suptitle('TotalBill과 Tips의 조인트차트')
plt.show()
```

### kedplot()

```python
kde, ax = plt.subplots()
sns.kdeplot( data=tips, x='total_bill', y='tip', fill=True, ax=ax)
ax.set_title('KDE Plot')
plt.show()
```

### barplot()

```python
kde, ax = plt.subplots()
sns.barplot( data=tips, x='time', y='total_bill', estimator=np.mean, ax=ax)
                                                  # y값이 mean
ax.set_title('BAR Plot')
plt.show()

# 세로선: 신뢰구간 (errorbar=None)
```

### boxplot()

```python
kde, ax = plt.subplots()
sns.boxplot( data=tips, x='time', y='total_bill')
ax.set_title('Box Plot')
plt.show()
```

### violinplot()
- 데이터 분포 보여줌
- hist + box

```python
kde, ax = plt.subplots()
sns.violinplot( data=tips, x='time', y='total_bill', ax=ax)
ax.set_title('violin Plot')
plt.show()
```

## 3.3 pairplot()

```python
fig = sns.pairplot(data=tips)
fig.figure.suptitle('Tips 데이터의 관계분석')
plt.show()
```
